```python
import param
import panel as pn
import pandas as pd
import sqlite3


class SQLWidget(param.Parameterized):
    query = param.String(default="SELECT * FROM table", doc="SQL Query")
    run_query = param.Action(lambda x: x.param.trigger('run_query'), label='Run Query')
    result = param.DataFrame()

    def __init__(self, **params):
        super().__init__(**params)
        self._conn = sqlite3.connect(':memory:')  # Connect to a SQLite database. Replace with your connection

    @param.depends('run_query', watch=True)
    def _run_query(self):
        self.result = pd.read_sql_query(self.query, self._conn)

    def view(self):
        return pn.Column(
            pn.Param(self.param.query, widgets={'query': pn.widgets.TextAreaInput}),
            pn.Param(self.param.run_query, widgets={'run_query': pn.widgets.Button(name='Run Query')}),
            pn.Param(self.param.result, widgets={'result': pn.widgets.DataFrame(width=600, height=400)})
        )


sql_widget = SQLWidget()
sql_widget.view().servable()
```
